<b>Task 3: Create a function to calculate the distance between two coordinates using the haversine formula.</b><br>The function haversine below will calculate a distance and return that distance(d). It takes as an input a pair of lists formatted [lat, long]. There are serveral ways to format this input, one of which is shown in the example below: the input is a pair of lists with the coordinates 30, 30, and 60, 60. It can also be formatted along the lines of<br><code>l1 = [30, 30]<br>l2 = [60, 60]<br>w=haversine(l1, l2)</code>

```python
def haversine(coor1, coor2):
    a = (math.sin((coor2[0] - coor1[0])/2))**2 + math.cos(coor1[0]) * math.cos(coor2[0]) * (math.sin((coor2[1] - coor1[1])/2))**2
    c = 2 * math.atan2(a**0.5, (1-a)**0.5)
    d = 6371 * c
    return d
w = haversine([30, 30], [60, 60])
print(w)
```

<b>Question 1: How many trees in the portland_park_trees dataset are of the Quercus or Acer genera?</b><br>There are multiple ways to answer this question, but I have demonstrated what I believe to be the simplest below. First, the dataset is loaded in as a Pandas dataframe using the function <code>pd.read_csv</code>. Then, a new dataframe is defined as a query of the first dataframe, returning a dataframe which only includes trees in the noted genera. Finally, the length of this dataframe is returned.

```python
portland_trees = pd.read_csv('/arcgis/home/Lab_Assignment2_GG3209/portland_park_trees.csv')
q1 = portland_trees.query('Genus=="Quercus" or Genus=="Acer"')
print(len(q1))
```

<b>Question 2: How many trees are of the Quercus or Acer genus and have a DBH larger than 50 inches?</b><br>There are two ways to query a dataset which will answer this question. Since we already have defined a dataframe of trees from the correct genera, we could simply query a subset of that which has a DBH greater than 50 inches. Alternatively, we could run a query on the main dataset which would return the same subset. I have demonstrated both ways below. Then, as above, the length of the dataframes is returned to answer the question.

```python
q2 = portland_trees.query('(Genus=="Quercus" or Genus=="Acer") and DBH>50') 
## This is the more complex way to do it which it involves querying the main dataframe again.
q2e = q1.query('DBH>50')
## This is a simple way to do it which relies on querying the previous dataframe of only Quercus and Acer trees.
print(len(q2))
print(len(q2e))
## As can be seen, both ways return the same result.
```

<b>Question 3: Which genus has the highest mean DBH of the following genera: Quercus, Acer, or Fraxinus?</b><br>As above, there are multiple possible ways to do this. I have demonstrated what are (in my opinion) the two simplest and cleanest ways. As can be seen, they return the same results. 

```python
## Method #1: create a series from a groupby query of the main dataset and then return from that series the values
## of each genus of interest.
tree_stats = portland_trees.groupby("Genus")['DBH'].mean()
print(tree_stats["Acer"])
print(tree_stats["Quercus"])
print(tree_stats["Fraxinus"])
## Method #2: create a new dataframe from a query of the main dataframe which only includes the genera that we are interested in
## then run a groupby query on this dataframe to create a series with the mean DBH values of the genera of interest. This
## method is convenient as it returns all three results from one command.
q3 = portland_trees.query('Genus=="Acer" or Genus=="Quercus" or Genus=="Fraxinus"')
q3_stats = q3.groupby("Genus")['DBH'].mean()
print(q3_stats)
```

<b>Question 4: How many different species of trees are recorded in the Acer genus?</b><br>I have identified two simple ways to do this. The first is to run a groupby on the portland_trees df with <code>nunique()</code> to identified the number of unique values per genera, the result of which is written to the series tree_spec. Then, the value for the Acer genus can be called. Alternatively, a new dataframe of only trees in the Acer genus can be created by querying the portland_trees dataframe, and then the number of unique values in the Genus_spec column identified in a similar way.

```python
## Method #1: finding unique species from the main dataframe.
tree_spec = portland_trees.groupby("Genus")['Genus_spec'].nunique()
print(tree_spec["Acer"])
# Method #2: finding unique species from a subset dataframe 'acer'. 
acer = portland_trees.query('Genus=="Acer"')
acer["Genus_spec"].nunique()
## As can be seen, the two methods return the same result.
```

<b>Graph 1: Create a scatterplot for just trees in the Ulmus genus with DBH mapped to the x-axis and tree height mapped to the y-axis.</b><br>There is more than one way to do this with the Seaborn graphing functions. Below, a method utilising the <code>relplot()</code> function is shown, but this could also be done with <code>sns.scatterplot()</code>, which would allow it to be used as a subplot. The code <code>[portland_trees["Genus"]=="Ulmus"]</code> is used to select a subset of the dataset where the genus is Ulmus.

```python
sns.relplot(x="DBH", y="TreeHeight", data=portland_trees[portland_trees["Genus"]=="Ulmus"])
```

<b>Graph 2: Create a scatterplot for just trees in the Ulmus genus with DBH mapped to the x-axis, tree
height mapped to the y-axis, and tree species mapped to hue.</b><br>The code for this is very similar to graph 1, with the change that the Genus_spec field is now mapped to the hue.

```python
sns.relplot(x="DBH", y="TreeHeight", hue="Genus_spec", data=portland_trees[portland_trees["Genus"]=="Ulmus"])
```

<b>Graph 3: Create a boxplot of DBH for just the Ulmus genus differentiated by species (or, each species should have its own boxplot).</b><br>Unlike the previous graphs, the command utilised in this is the <code>sns.boxplot</code> function. This is more analogous to <code>sns.scatterplot()</code> as it can be used as a subplot.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=portland_trees[portland_trees["Genus"]=="Ulmus"], x='Genus_spec', y='DBH')
plt.title('DBH by Species in the Ulmus Genus')
plt.show()
```

<b>Graph 4: Combine Graphs 1 and 3 into a single figure.</b><br>This was accomplished using <code>plt.subplot</code>, specifying (1, 2) to create a figure with 2 columns and 1 row. Note that <code>sns.scatterplot()</code> is used instead of <code>sns.relplot()</code>, as the latter is incompatible with the subplot function.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4))
sns.scatterplot(ax=ax1, x="DBH", y="TreeHeight", data=portland_trees[portland_trees["Genus"]=="Ulmus"])
sns.boxplot(ax=ax2, data=portland_trees[portland_trees["Genus"]=="Ulmus"], x='Genus_spec', y='DBH')
```
